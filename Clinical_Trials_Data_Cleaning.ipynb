```python
# fill the null values
patients.address.fillna('unknown',inplace = True)
patients.city.fillna('unknown',inplace = True)
patients.state.fillna('unknown',inplace = True)
patients.country.fillna('unknown',inplace = True)

# find and fill null values of zipcode
patients[patients['zip_code'].isna()]
patients.zip_code.fillna(0,inplace = True)
```

```python
# verifying the null values in zip_code
patients.zip_code.isna().sum()
```

#### Zip_code Datatype is in Float, it should be in Int

```python
# converting the zip_code datatype from float to int
patients.zip_code = patients.zip_code.astype(int)
```

#### Contact Column has mixed informatation of email and mobile numbers of customers,so i am trying to separate the email and mobile numbers of customers and fill them into a new colums for each  

```python
# extracting the email and phone number from contact
phone_pattern = r'(\+?\d{1,4}[\s-]?\(?\d{1,4}\)?[\s-]?\d{1,4}[\s-]?\d{1,9})'

patients['phone_number'] = patients['contact'].str.extract(phone_pattern)
patients['phone_number'] = patients['phone_number'].str.replace('[\(\)\+\-\s]', '', regex=True)

# filling the nan values with 0
patients['phone_number'].fillna('0', inplace=True)

#converting datatype of the phone_number to str
patients['phone_number'] = patients['phone_number'].astype(str)

#removing the countrycode and other digits from phone_number
def phone_number(number):
    if len(number) == 11:
        number = number[1:]
    return number

patients['phone_number'] = patients['phone_number'].apply(phone_number)
patients['phone_number'].head()
```

```python
# extracting the email from contact
email_pattern = r'([a-zA-Z._%+-]+@[a-zA-Z.-]+\.[a-zA-Z]{2,})'
patients['email'] = patients['contact'].str.extract(email_pattern)
# filling null values with Unknown
patients['email'].fillna('unknown',inplace = True)
```

#### Verifying the the email and mobile numbers whether they are separated

```python
patients[['contact', 'email', 'phone_number']].head(20)
```

```python
# removing the patients contact column
patients.drop('contact',axis = 1, inplace = True)
```

#### The values in weight column is mentioned in pound, but they should be in kg's
#### The Values in Height columns are mentioned in inches, but they should be in meters's

```python
#converting the weight from pounds to kg's
patients['weight'] = patients['weight'] * 0.453592

# converting the height from inches to mt's
patients['height'] = patients['height'] * 0.0254
```

#### The bmi column values are incorrectly calculated,so i am trying to calculate the correct bmi with revised weight's and Height's of the customer's
##### bmi: the Body Mass Index (BMI) of each patient. BMI is a simple calculation using a person's height and weight. The formula is BMI = kg/m2 where kg is a person's weight in kilograms and m2 is their height in metres squared.

```python
#BMI = kg/m2 calculation and creating column
patients['bmi'] = patients['weight'] / (patients['height'] **2)
```

```python
patients
```

#### Calculating the bmi_status using the bmi values
##### A BMI of 25.0 or more is overweight, while the healthy range is 18.5 to 24.9. The inclusion criteria for this clinical trial is 16 >= BMI >= 38.


```python
def status(bmi):
  if bmi < 18.5:
      return 'Underweight'
  elif 18.5 <= bmi < 24.9:
      return 'Normal weight'
  elif 24.9 <= bmi < 29.9:
      return 'Overweight'
  elif 29.9 <= bmi < 34.9:
      return 'Obesity'
  else:
      return 'Severe Obesity'

# Apply the function to each value in the 'revised_bmi' column
Bmi_Status = patients['bmi'].apply(status)
```

```python
# creating the Bmi_status column
patients['bmi_status'] = Bmi_Status
patients['bmi_status']
```

#### The Baseline and end dose of auralin and novodra are in units and the change in the dose from baseline(start) and end was not mentioned. here i'm' calculating the change for auraline and novodra dose's

```python
# Extracting the start and end doses for Auralin
auralin_start = treatments['auralin'].str.extract(r'(\d+)u -')[0]
auralin_end = treatments['auralin'].str.extract(r'- (\d+)u')[0]

# Converting the extracted values to numeric
treatments['auralin_start'] = pd.to_numeric(auralin_start, errors='coerce')
treatments['auralin_end'] = pd.to_numeric(auralin_end, errors='coerce')
```

```python
# Extracting the start and end doses for Novodra
novodra_start = treatments['novodra'].str.extract(r'(\d+)u -')[0]
novodra_end = treatments['novodra'].str.extract(r'- (\d+)u')[0]

# Converting the extracted values to numeric
treatments['novodra_start'] = pd.to_numeric(novodra_start, errors='coerce')
treatments['novodra_end'] = pd.to_numeric(novodra_end, errors='coerce')
```

```python
treatments['start_dose'] = treatments['auralin_start'].combine_first(treatments['novodra_start'])
treatments['end_dose'] = treatments['auralin_end'].combine_first(treatments['novodra_end'])
```

```python
treatments['drug_type'] = treatments.apply(lambda row: 'Auralin' if pd.notna(row['auralin_start']) else 'Novodra', axis=1)
```

#### The hba1c_change column values are incorrectly calculated, so i'm re calculating the hba1c_change values.hba1c_change = hba1c_start - hba1c_end

```python
treatments['hba1c_change'] = treatments['hba1c_start'] - treatments['hba1c_end']
```
